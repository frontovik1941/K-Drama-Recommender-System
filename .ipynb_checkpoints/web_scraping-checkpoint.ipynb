```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
from pandas import Series,DataFrame
```

```python
pages = []

#33 pages of different kdramas
for i in range(1,34):
    url = 'https://mydramalist.com/search?adv=titles&ty=68&co=3&nt=1,2,3,4,6,33,37,51,128&re=2012,2022&rt=7.5,10&st=3&so=rated&page=' + str(i)
    result = requests.get(url)
    page = BeautifulSoup(result.content)
    pages.append(page)
```

```python
#list for all the individual kdrama page links
#these pages include descriptions/features about the drama
#also grabs date released

indiv_links= []
Dates = []

for page in pages:
    tot_list = page.find_all("h6",{'class': 'text-primary title'})
    for link in tot_list:
        indiv_links.append(link.find('a',href=True)['href'])
    date_list = page.find_all("div",{'class': 'col-xs-9 row-cell content'})
    for date in date_list:
        tmp = date.find("span",{'class':'text-muted'}).get_text().strip()
        Dates.append(int(tmp[15:19]))
```

```python
#open each link and get features

#list for features
Name = []
Ep_Count = []
Age_Rating = []
Cast = []
Network = []
Genre = []
Score = []
Tags = []

for link in indiv_links:
    url = 'https://mydramalist.com' + link
    result = requests.get(url)
    data = BeautifulSoup(result.content)
    name = data.find('span',{'itemprop': 'name'}).get_text()
    Name.append(name)
    age = data.find('li',{'class':'list-item p-a-0 content-rating'})
    age.find('b',{'class': 'inline'}).decompose()
    Age_Rating.append(age.get_text().strip())

    list_items = data.find('div',{'class':'box clear hidden-sm-down'}).find_all('li',{'class':'list-item p-a-0'})
    for item in list_items:
        item.find('b',{'class': 'inline'}).decompose()
    
    cast_list = []
    casts = data.find_all('a',{'class':'text-primary text-ellipsis'})
    
    for cast in casts:
        cast_list.append(cast.find('b',{'itempropx':'name'}).get_text().strip())
    Cast.append(cast_list)

    Ep_Count.append(int(list_items[2].get_text()))
    Network.append(list_items[5].get_text().strip())

    genres = data.find('li',{'class':'list-item p-a-0 show-genres'}).find_all('a')
    for i in range(len(genres)):
        genres[i] = genres[i].get_text().strip()
    Genre.append(genres)

    tags = data.find('li',{'class':'list-item p-a-0 show-tags'}).find_all('a')
    for i in range(len(tags)):
        tags[i] = tags[i].get_text().strip()
    Tags.append(tags)

    score = data.find('b',{'style':'font-weight:bold;','itempropx':'ratingValue'})
    Score.append(float(score.get_text()))
```

```python
features_df = pd.concat([Series(Name),Series(Dates),Series(Ep_Count),Series(Age_Rating),Series(Cast),
                         Series(Network),Series(Genre),Series(Tags),Series(Score)],axis=1)
features_df.head()
```

```python
features_df.info()
```

```python
features_df.columns = ['Name','Date','Ep_Count','Age_Rating','Cast','Network','Genres','Tags','Score']
```

```python
features_df.Age_Rating.value_counts()
```

```python
def clean_age_rating(rating):
    if(rating != 'Not Yet Rated'):
        if(rating == 'G - All Ages'):
            return '13+'
        return rating[:3]
    return rating
features_df['Age_Rating'] = features_df['Age_Rating'].apply(clean_age_rating) 
```
